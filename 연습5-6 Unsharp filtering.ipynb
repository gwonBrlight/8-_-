```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image-processing/img1.bmp')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#edge_img = np.zeros((img.shape[0],img.shape[1]))

# 3x3커널 적용하기
mean_img_3by3 = cv2.blur(gray_img,(3,3))

edge_img_3by3 = cv2.addWeighted(gray_img, 1.0, mean_img_3by3, -1.0, 0)
output_img_3by3 = cv2.addWeighted(gray_img, 1.0, edge_img_3by3, 3.0, 0)

# 7x7커널 적용하기
mean_img_7by7 = cv2.blur(gray_img,(7,7))

edge_img_7by7 = cv2.addWeighted(gray_img, 1.0, mean_img_7by7, -1.0, 0)
output_img_7by7 = cv2.addWeighted(gray_img, 1.0, edge_img_7by7, 3.0, 0)

# 15x15커널 적용하기
mean_img_15by15 = cv2.blur(gray_img,(15,15))

edge_img_15by15 = cv2.addWeighted(gray_img, 1.0, mean_img_15by15, -1.0, 0)
output_img_15by15 = cv2.addWeighted(gray_img, 1.0, edge_img_15by15, 3.0, 0)


#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(1,3,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,3,2)
plt.title("Edge Image_3by3")
plt.imshow(edge_img_3by3, cmap='gray')
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Unsharped Image_3by3")
plt.imshow(output_img_3by3, cmap='gray')
plt.axis("off")
plt.show()

plt.figure(figsize=(20,20))
plt.subplot(1,3,2)
plt.title("Edge Image_7by7")
plt.imshow(edge_img_7by7, cmap='gray')
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Unsharped Image_7by7")
plt.imshow(output_img_7by7, cmap='gray')
plt.axis("off")
plt.show()

plt.figure(figsize=(20,20))
plt.subplot(1,3,2)
plt.title("Edge Image_15by15")
plt.imshow(edge_img_15by15, cmap='gray')
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Unsharped Image_15by15")
plt.imshow(output_img_15by15, cmap='gray')
plt.axis("off")
plt.show()


cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/unsharpfilter_3by3_img1.bmp',output_img_3by3)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/unsharpfilter_7by7_img1.bmp',output_img_7by7)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/unsharpfilter_15by15_img1.bmp',output_img_15by15)
```
