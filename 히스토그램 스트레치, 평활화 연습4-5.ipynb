```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
src = cv2.imread('/content/gdrive/My Drive/Image-processing/baboon.jpg', cv2.IMREAD_GRAYSCALE)
streched = cv2.normalize(src, None, 0, 255, cv2.NORM_MINMAX)
equalize = cv2.equalizeHist(src)

plt.figure(figsize=(15,5))
plt.subplot(131), plt.imshow(src, 'gray'), plt.title('src1')
plt.subplot(132), plt.imshow(streched, 'gray'), plt.title('Streched')
plt.subplot(133), plt.imshow(equalize, 'gray'), plt.title('Equalize')
plt.show()

hist = cv2.calcHist([src], [0], None, [256], [0, 256]) 
output_img_hist = cv2.calcHist([streched], [0], None, [256], [0, 256]) 
output_equalizeHist = cv2.calcHist([equalize], [0], None, [256], [0, 256]) 

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title("Histogram")
plt.plot(hist)

plt.subplot(1,3,2)
plt.title("Streched Histogram")
plt.plot(output_img_hist)

plt.subplot(1,3,3)
plt.title("Equalize Histogram")
plt.plot(output_equalizeHist)
plt.show()
```
