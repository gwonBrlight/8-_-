```python
from google.colab import drive
drive.mount('/content/gdrive')
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
from scipy import signal
from math import log10, sqrt
import matplotlib.pylab as pylab
import numpy.fft as fp
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/gdrive/MyDrive/Image-processing/Copy of fce(salt_pepper noise).bmp', cv2.IMREAD_GRAYSCALE)
img = cv2.imread("/content/gdrive/MyDrive/Image-processing/Copy of Lena-Gaussian-noise2.jpg", cv2.IMREAD_GRAYSCALE)


def LPF(img):

  freq = fp.fft2(img)
  gauss_kernel = np.outer(signal.gaussian(img.shape[0], 1.5), signal.gaussian(img.shape[1], 1.5))
  assert(freq.shape == gauss_kernel.shape)
  freq_kernel = fp.fft2(fp.ifftshift(gauss_kernel))
  #print((freq_kernel))
  convolved = (freq * freq_kernel)
  out_img = fp.ifft2(convolved).real
  return out_img

resultImg_man = LPF(img1)
resultImg_lena = LPF(img)
resultImg_man = cv2.normalize(resultImg_man, None, 0, 255, cv2.NORM_MINMAX)
resultImg_lena = cv2.normalize(resultImg_lena, None, 0, 255, cv2.NORM_MINMAX)

plt.figure(figsize=(30,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Salt-and-peper")
plt.imshow(img1, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("LPF_Salt-and-peper")
plt.imshow(resultImg_man, cmap='gray')
plt.axis("off")
plt.show()

plt.figure(figsize=(30,20))# 영상의 크기를 키워주자
plt.subplot(1,2,1)
plt.title("Gaussian-noise")
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("LPF_Gaussian-noise")
plt.imshow(resultImg_lena, cmap='gray')
plt.axis("off")
plt.show()
```
