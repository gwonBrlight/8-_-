```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image-processing/img3.bmp')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 3x3 중간값 커널 적용하기
output_img_3by3 =cv2.GaussianBlur(gray_img,(3,3),1)
# 7x7 중간값 커널 적용하기
output_img_7by7 = cv2.GaussianBlur(gray_img,(7,7),1)
# 15x15 중간값 커널 적용하기
output_img_15by15 = cv2.GaussianBlur(gray_img,(15,15),1)

#그림을 화면에 출력
plt.figure(figsize=(30,20))# 영상의 크기를 키워주자
plt.subplot(1,4,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,4,2)
plt.title("gaussian filtering Image_3by3")
plt.imshow(output_img_3by3, cmap='gray')
plt.axis("off")
plt.subplot(1,4,3)
plt.title("gaussian filtering Image_7by7")
plt.imshow(output_img_7by7, cmap='gray')
plt.axis("off")
plt.subplot(1,4,4)
plt.title("gaussian filtering Image_15by15")
plt.imshow(output_img_15by15, cmap='gray')
plt.axis("off")
plt.show()

cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/gaussianfilter_3by3_img3.bmp',output_img_3by3)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/gaussianfilter_7by7_img3.bmp',output_img_7by7)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/gaussianfilter_15by15_img3.bmp',output_img_15by15)
```
