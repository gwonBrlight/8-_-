```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img_color = cv2.imread('/content/gdrive/My Drive/Image-processing/baboon.jpg')
# BGR채널순서를 RGB채널로 변경

img_r,img_g,img_b = cv2.split(img_color)

zeros = np.zeros((img_color.shape[0], img_color.shape[1]), dtype="uint8")

img_r = cv2.merge([img_r, zeros, zeros])
img_g = cv2.merge([zeros, img_g, zeros])
img_b = cv2.merge([zeros, zeros, img_b])

R_output_img=np.zeros((img_r.shape[0],img_r.shape[1]))
G_output_img=np.zeros((img_g.shape[0],img_g.shape[1]))
B_output_img=np.zeros((img_b.shape[0],img_b.shape[1]))

for h in range(img_b.shape[0]):
  for w in range(img_b.shape[1]):
    if(np.int32(img_b[h,w][2])>100):
        B_output_img[h,w]=255
    else:
        B_output_img[h,w]=0

for h in range(img_g.shape[0]):
  for w in range(img_g.shape[1]):
    if(np.int32(img_g[h,w][1])>100):
        G_output_img[h,w]=255
    else:
        G_output_img[h,w]=0

for h in range(img_r.shape[0]):
  for w in range(img_r.shape[1]):
    if(np.int32(img_r[h,w][0])>100):
        R_output_img[h,w]=255
    else:
        R_output_img[h,w]=0



plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.title("B_output_img")
plt.imshow(img_b)
plt.subplot(1,3,2)
plt.title("G_output_img")
plt.imshow(img_g)
plt.subplot(1,3,3)
plt.title("R_output_img")
plt.imshow(img_r)

plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.title("B_output_img, 임곗값 100")
plt.imshow(B_output_img)
plt.subplot(1,3,2)
plt.title("G_output_img, 임곗값 100")
plt.imshow(G_output_img)
plt.subplot(1,3,3)
plt.title("R_output_img, 임곗값 100")
plt.imshow(R_output_img)


plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.title("B_output_img, 임곗값 100")
plt.imshow(B_output_img,cmap='gray')
plt.subplot(1,3,2)
plt.title("G_output_img, 임곗값 100")
plt.imshow(G_output_img,cmap='gray')
plt.subplot(1,3,3)
plt.title("R_output_img, 임곗값 100")
plt.imshow(R_output_img,cmap='gray')

```
