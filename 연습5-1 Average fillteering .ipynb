```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image-processing/img3.bmp')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 계수가 1로 구성된 3x3커널 만들기
kernel_3 = np.ones((3,3),np.float32)/9
output_img_3by3 = cv2.filter2D(gray_img,-1,kernel_3)

# 계수가 1로 구성된 7x7커널 만들기
kernel_7 = np.ones((7,7),np.float32)/49
output_img_7by7 = cv2.filter2D(gray_img,-1,kernel_7) 

# 계수가 1로 구성된 15x15커널 만들기
kernel_15 = np.ones((15,15),np.float32)/225
output_img_15by15 = cv2.filter2D(gray_img,-1,kernel_15) 

#그림을 화면에 출력
plt.figure(figsize=(30,20))# 영상의 크기를 키워주자
plt.subplot(1,4,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Average filtering Image_3by3")
plt.imshow(output_img_3by3, cmap='gray')
plt.axis("off")
plt.subplot(1,4,3)
plt.title("7by7_Average filtering Image_")
plt.imshow(output_img_7by7, cmap='gray')
plt.axis("off")
plt.subplot(1,4,4)
plt.title("Average filtering Image_15by15")
plt.imshow(output_img_15by15, cmap='gray')
plt.axis("off")
plt.show()

cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/averagefilter_3by3_img3.bmp',output_img_3by3)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/averagefilter_7by7_img3.bmp',output_img_7by7)
cv2.imwrite('/content/gdrive/My Drive/Image-processing/result/averagefilter_15by15_img3.bmp',output_img_15by15)
```
